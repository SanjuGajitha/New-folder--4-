```python
# get the dataset values in to array
all_data = df.values
print(all_data)
```

```python
print(all_data.shape)
```

```python
#divide the data sets in to data and target
print(df.columns)
```

```python
#divide the all data set in to data
data=all_data[:,0:20]
print(data.shape)
```

```python
#divide the all data set in to target
target=all_data[:,20]
print(target.shape)
```

```python
#split data and target in to train and test data
from sklearn.model_selection import train_test_split
train_data,test_data,train_target,test_target=train_test_split(data,target,test_size=0.2)
print(test_data.shape)
```

```python
#using SVM make model and train data
from sklearn.svm import SVC
model=SVC(kernel='rbf')
model.fit(train_data,train_target)

```

```python
#test data
from sklearn.metrics import classification_report, accuracy_score
predicted_data=model.predict(test_data)
print(accuracy_score(test_target,predicted_data))
```

```python
print("Classification Report:")
print(classification_report(test_target, predicted_data))
```

```python
import joblib

# Save the model to a file
joblib.dump(model, 'svm_phone_price_model.pkl')
```
